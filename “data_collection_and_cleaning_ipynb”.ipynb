```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#airbnb data
!wget https://storage.googleapis.com/public-data-data_viz-2/london-listings-master-2019.csv
!wget https://storage.googleapis.com/public-data-data_viz-2/newyork-listings-master-2019.csv
!wget https://storage.googleapis.com/public-data-data_viz-2/paris-listings-master-2019.csv
```

```python
london_airbnb = pd.read_csv('london-listings-master-2019.csv')
newyork_airbnb = pd.read_csv('newyork-listings-master-2019.csv')
paris_airbnb = pd.read_csv('paris-listings-master-2019.csv')
```

```python
print(london_airbnb.shape, newyork_airbnb.shape, paris_airbnb.shape)
```

```python
london_airbnb.isna().sum()
```

```python
london_airbnb.dropna()
```

```python
london_airbnb.describe()
```

```python
london_airbnb['price_float'] = london_airbnb['price'].str.replace(',', '').astype(float)
```

```python
london_airbnb.describe()
```

```python
mean = london_airbnb['price_float'].mean()
mean
```

```python
std = london_airbnb['price_float'].std()
std
```

```python
london_airbnb.price_float
```

```python
london_airbnb_1 = london_airbnb[london_airbnb['price_float']<=880]
london_airbnb_1
```

```python
london_airbnb_1.dropna()
```

```python
london_airbnb_1['last_review'] = pd.to_datetime(london_airbnb_1['last_review'])
```

```python
london_airbnb_1.sort_values('last_review', inplace=True)
```

```python
london_airbnb_1.head()
```

```python
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('last_review', 'price_float', data=london_airbnb_1, color='tab:red')
```

```python
newyork_airbnb['price_float'] = newyork_airbnb['price'].str.replace(',', '').astype(float)
```

```python
mean = newyork_airbnb['price_float'].mean()
std = newyork_airbnb['price_float'].std()
mean
```

```python
s = mean+3*std
s
```

```python
newyork_airbnb_1 = newyork_airbnb[newyork_airbnb['price_float']<=s]
newyork_airbnb_1
```

```python
newyork_airbnb_1.dropna()
```

```python
newyork_airbnb_1['last_review'] = pd.to_datetime(newyork_airbnb_1['last_review'])
```

```python
newyork_airbnb_1.sort_values('last_review', inplace=True)
```

```python
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('last_review', 'price_float', data=newyork_airbnb_1, color='tab:red')
```

```python
paris_airbnb['price_float'] = paris_airbnb['price'].str.replace(',', '').astype(float)
```

```python
mean = paris_airbnb['price_float'].mean()
std = paris_airbnb['price_float'].std()
mean
```

```python
s = mean+3*std
s
```

```python
paris_airbnb_1 = paris_airbnb[paris_airbnb['price_float']<=s]
paris_airbnb_1
```

```python
paris_airbnb_1.dropna()
```

```python
paris_airbnb_1['last_review'] = pd.to_datetime(paris_airbnb_1['last_review'])
```

```python
paris_airbnb_1.sort_values('last_review', inplace=True)
```

```python
plt.figure(figsize=(16,10), dpi= 80)
plt.plot('last_review', 'price_float', data=paris_airbnb_1, color='tab:red')
plt.rcParams['agg.path.chunksize'] = 10000
```
